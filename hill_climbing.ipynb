We define a number of helper functions used for plotting and analysis.

```python
def orthonormalize(w):
    """Returns an orthonormalized version of a matrix (w)."""
    return w.dot(inv(sqrtm(w.T.dot(w)))).real

def random_basis(d):
    """Returns a randomized d-dimensional basis."""
    b = UniformHypersphere().sample(d, d)
    return orthonormalize(b)

def xyz(func, res):
    """Returns a meshgrid (x, y, z), where z = func(x, y)."""
    lx, ly = np.linspace(-1, 1, res), np.linspace(1, -1, res)
    x, y = np.meshgrid(lx, ly)
    z = func(x, y)
    return x, y, z

def func_min(func, res=2001):
    """Returns the (x, y) point which obtains the minimum of func."""
    x, y, z = xyz(func, res)
    xy = np.argmin(z.flatten())
    return 2 * (xy % len(z)) / float(res - 1) - 1, 1 - 2 * (xy / len(z)) / float(res - 1)

def plot_func(ax, func, res=101, num_ticks=5, num_samples=10000, **kwargs):
    """Plot the function (func) to the axis (ax) as a 2D heatmap.""" 
    cmap = sns.cubehelix_palette(start=0, light=1.0, reverse=True, as_cmap=True)
    cmap_contour = sns.cubehelix_palette(start=0, light=0.7, dark=-0.1, reverse=False, as_cmap=True) 

    x, y, z = xyz(func, res)
    
    lx, ly = np.linspace(-1, 1, res), np.linspace(1, -1, res)
    ticksl = np.linspace(0, res-1, num_ticks).astype('int')
    xticks = np.asarray(['    ']*res)
    xticks[ticksl] = lx[ticksl]
    yticks = np.asarray(['    ']*res)
    yticks[ticksl] = ly[ticksl]

    vmin, vmax = z.min(), z.max()
    argmin = z.argmin()
    sns.heatmap(z.copy(), vmin=vmin, vmax=vmax, center=(vmin+vmax)/2, cmap=cmap, 
                robust=False, annot=False, linewidths=0, square=True,
                xticklabels=xticks, yticklabels=yticks, ax=ax, **kwargs)
    ax.set_yticklabels(yticks[::-1], rotation=0)  # otherwise they will be rotated 90 degrees

def plot_3d(ax, func, label, res=101, elev=20, azim=35):
    """Plots the function (func) to the axis (ax) in 3D."""
    x, y, z = xyz(func, res)
    ax.plot_surface(x, y, z, rstride=1, cstride=1, 
                    cmap=sns.cubehelix_palette(as_cmap=True, light=0.1, dark=1.0),
                    linewidth=0, antialiased=True, shade=False)
    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel(r'${\bf x_0}$')
    ax.set_ylabel(r'${\bf x_1}$')
    ax.zaxis.set_rotate_label(False)
    ax.set_zlabel(r'  $f_%d({\bf x})$' % label, rotation=0)
```

The test functions $f_1$, $f_2$, $f_3$, as given in the paper:

```python
def func1(x, y):
    return ((x + 0.3)**2 + (y - 0.4)**2 - 2)*0.5

def func2(x, y):
    return 0.2*(np.sin(6*x) + np.sin(6*y) - np.sin(2*x)*np.sin(4*y)) + 0.4*(x + 0.3)**2 + 0.4*(y - 0.5)**2 - 0.5

def func3(x, y):
    return 0.7*(-.3 - np.exp(-3*(abs(x + y))) + ((x - y) / 2 - 0.6)**2)

# make sure f3 is an actual ridge
assert func3(0.001, 0) > func3(0, 0)
assert func3(0, 0.001) > func3(0, 0)
```

### Model Creation

Define all constants used for the model.

```python
# Model parameters
period = 0.2        # period of pulses
pulse = 0.02        # duration of pulses
pulse_scaler = 10   # weighting on pulse before normalization
lookback = 0.01     # interval of differentation
dt = 0.001          # simulation timestep

# Neural parameters
scale_a = 50         # scaling on acceleration
scale_v = 5          # scaling on velocity
tau_v = 0.005        # synapse tau on velocity
tau_x = 0.005        # synapse tau on position
tau_feedback = 0.01  # synapse tau on feedback
tau_probe = 0.002    # synapse tau on probes
```

Create the Nengo object that will generate the appropriate error signal driving acceleration.

```python
class Intuition(object):
    """Feedback loop, implemented without using Neurons."""
    
    def __init__(self, f, dims, period, pulse, pulse_scaler, lookback, dt, basis=None):
        self._buff = []
        self._elapsed = 0
        self.f = f
        self.dims = dims
        self.period = period
        self.pulse = pulse
        self.pulse_scaler = pulse_scaler
        self.lookback = lookback
        self.dt = dt
        self.basis = random_basis(dims) if basis is None else np.asarray(basis)

    def __call__(self, t, x):
        # Keep buffer of time length `lookback`
        self._buff.append(x.copy())
        if len(self._buff) > self.lookback/self.dt:
            self._buff = self._buff[1:]
            
        # Pulsate for time `self.pulse` each `self.period`
        self._elapsed += self.dt
        if self._elapsed >= self.period:
            self._elapsed = 0
        if 0 <= self._elapsed <= self.pulse:
            i = int(t / self.period + 1e-4) % self.dims
            z = self.basis[i, :]
        else:
            z = 0

        # Force in the direction of velocity, flipped if error
        dx = self._buff[-1] - self._buff[0]
        dy = self.f(*self._buff[-1]) - self.f(*self._buff[0])
        out = -dx*dy + z*self.pulse_scaler
        if np.linalg.norm(out) < 1e-4:  # epislon
            return np.zeros(self.dims)  # zero-division
        return out / np.linalg.norm(out)
```

Plot the above signal given no input, to generate `u(t)` for 5 periods. Save to **`u.png`**

```python
def plot_intuition(periods):
    """Plots the intuition given no input."""
    dims = 2
    intuition = Intuition(
        f=lambda *args, **kwargs: 0, dims=dims, period=period, pulse=pulse, 
        pulse_scaler=pulse_scaler, lookback=lookback, dt=dt, basis=np.eye(dims))
    t = np.arange(dt, dims*period*periods, dt)
    y = filtfilt([intuition(t_i, np.zeros(dims)) for t_i in t], tau_v/dt)
    with sns.plotting_context('paper'):
        fig = plt.figure(figsize=(10, 2))
        plot(t, y)
        xlabel('Time ($s$)')
        ylabel(r'${\bf u}(t)$', rotation=0, labelpad=15)
        show()
        fig.savefig('u.png', bbox_inches='tight', dpi=300)
    
plot_intuition(periods=5)
```

Below is the code that will run a single simulation trial for a variable number of neurons and time. 

Decoded activity and spike trains for each of the three populations are returned.

```python
def run(f, n_neurons=1000, dims=2, dt=0.001, T=3.0):
    with nengo.Network() as model:
        # State
        intuition = nengo.Node(
            size_in=dims, output=Intuition(
                f=f, dims=dims, period=period, pulse=pulse, 
                pulse_scaler=pulse_scaler, lookback=lookback, dt=dt))
        x = nengo.Ensemble(n_neurons, dims)
        v = nengo.Ensemble(n_neurons, dims) 
        a = nengo.Ensemble(n_neurons, dims) 

        # Project to intuition
        nengo.Connection(x, intuition, synapse=tau_feedback)

        nengo.Connection(intuition, a, synapse=tau_v, transform=scale_a)

        # Update velocity
        nengo.Connection(a, v, synapse=tau_v, transform=tau_v)
        nengo.Connection(v, v, synapse=tau_v)

        # Update position
        nengo.Connection(v, x, synapse=tau_x, transform=tau_x*scale_v)
        nengo.Connection(x, x, synapse=tau_x)

        # Probes
        p_a = nengo.Probe(a, synapse=tau_probe)
        p_a_spikes = nengo.Probe(a.neurons, 'spikes')
        p_v = nengo.Probe(v, synapse=tau_probe)
        p_v_spikes = nengo.Probe(v.neurons, 'spikes')
        p_x = nengo.Probe(x, synapse=tau_probe)
        p_x_spikes = nengo.Probe(x.neurons, 'spikes')
        
    sim = nengo.Simulator(model, dt=dt)
    sim.run(T)
    
    return sim.trange(), sim.data[p_a], sim.data[p_v], sim.data[p_x], sim.data[p_a_spikes], sim.data[p_v_spikes], sim.data[p_x_spikes]
```

### Model Evaluation

This code generates the **`probes.png`** figure by running a single simulation and plotting all of the probed data.

```python
def add_spikes(ax, t, spikes, num_show=120, color_cycle=8, alpha=0.5, lw=1):
    """Adds spiking data of some of the neurons to the background of a plot."""
    ymin, ymax = ax.get_ylim()
    h = (ymax - ymin) / float(num_show)
    colors = sns.color_palette("Pastel1", color_cycle)  # husl
    for i in range(num_show):
        ax.vlines(t[spikes[:, i] > 0], ymin+i*h, ymin+(i+1)*h, 
                  color=colors[i % color_cycle], alpha=alpha, lw=lw)
    ax.set_ylim(ymin, ymax)

def plot_simdata(f):
    """Plots a single trial's simulation data."""
    t, a, v, x, asp, vsp, xsp = run(f)
    with sns.axes_style("white"):
        with sns.plotting_context('paper'):
            fig, axes = plt.subplots(3, 1, figsize=(6, 9), sharex=True)
            
            axes[2].plot(t, a)
            add_spikes(axes[2], t, asp)
            axes[2].set_ylabel(r'${\bf a}(t)$', rotation=0, labelpad=15)
            axes[2].set_title('Acceleration')
            
            axes[1].plot(t, v)
            add_spikes(axes[1], t, vsp)
            axes[1].set_ylabel(r'${\bf v}(t)$', rotation=0, labelpad=15)
            axes[1].set_title('Velocity')
            
            axes[0].plot(t, x)
            add_spikes(axes[0], t, xsp)
            axes[0].set_ylabel(r'${\bf x}(t)$', rotation=0, labelpad=15)
            axes[0].set_title('Position')
            plt.xlabel('Time ($s$)')
            
            sns.despine(left=True, bottom=True)
            show()
            fig.savefig('probes.png', bbox_inches='tight', dpi=300)
            
np.random.seed(50)
plot_simdata(func2)
```

Generate samples of paths taken by some arbitrarily chosen simulation on each function. Saves the results to **`paths.png`**

```python
def plot_path(ax, f, res=201, **kwargs):
    """Overplays the plot taken by a trial on a heatmap."""
    t, a, v, x, asp, vsp, xsp = run(f)
    colors = sns.cubehelix_palette(len(t), start=.2, dark=0.5, rot=-.55, reverse=True)
    plot_func(ax, f, res=res, **kwargs)
    ax.scatter(*zip(*(x+1)*res/2), c=colors, alpha=0.6, marker='o', s=20)
    
# Seeding in order to pick "interesting" samples
with sns.axes_style('white'):
    with sns.plotting_context('paper'):
        fig, axes = plt.subplots(1, 3, figsize=(22, 5))
        
        np.random.seed(25)
        axes[0].set_title(r'$f_1({\bf x})$')
        plot_path(axes[0], func1, cbar=False)
        
        np.random.seed(123)
        axes[1].set_title(r'$f_2({\bf x})$')
        plot_path(axes[1], func2, cbar=False)
        
        np.random.seed(100)
        axes[2].set_title(r'$f_3({\bf x})$')
        plot_path(axes[2], func3, cbar=True)
        
        show()
        fig.savefig('paths.png', bbox_inches='tight', dpi=300)
```

The following outputs the LaTex table used to quantitatively evaluate the method, with 100 trials on each function.

The output may be copy-pasted directly into the paper.

```python
def evaluate(f, trials, label, T=3):
    """Evaluate a number of trials on f lasting T seconds."""
    opt_x = func_min(f)
    opt = f(*opt_x)
        
    errors = []
    times = []
    for _ in range(trials):
        t, a, v, x, asp, vsp, xsp = run(f, T=T)

        y = np.asarray([f(*x_t) for x_t in x])
        opt_approx = y.min()
        opt_approx_t = t[y.argmin()]
        
        errors.append(opt_approx - opt)
        times.append(opt_approx_t)
        
    errors = np.asarray(errors)
    times = np.asarray(times)
    m_errors = np.mean(errors)
    m_times = np.mean(times)
    
    print r"$f_%d$ & $%.4f$ $(\pm%.3f)$ & $%.2fs$ $(\pm%.1fs)$ \\" % (
        label, m_errors, rmse(errors, m_errors), m_times, rmse(times, m_times))
        
trials = 100
evaluate(func1, trials, 1)
evaluate(func2, trials, 2)
evaluate(func3, trials, 3)
```
